```python
# Fazendo o fillna dos casos baseado na feature de landing_page
# Posterior serÃ¡ eliminada a feature landing_page

lista_fill = [
    '/emprestimos/garantia-veiculo/solicitar',
    '/emprestimos/garantia-veiculo/guia-bolso',
    '/blog/artigos/emprestimos/emprestimo-com-garantia-de-veiculo-alienado'
]

df['landing_page_product'] = list(
    map(
        lambda x,y: 'GarantiaVeiculo' if y in lista_fill else x,
        df['landing_page_product'], df['landing_page']
    )
) ; df['landing_page_product'].isna().sum()
```
